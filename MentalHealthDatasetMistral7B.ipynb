```python
from datasets import load_dataset
import re
```

```python
# Load the dataset
dataset = load_dataset('heliosbrahma/mental_health_chatbot_dataset')

# Shuffle the dataset and slice it
dataset = dataset['train'].shuffle(seed=42)
```

```python
def transform_conversation(example):
    # Extract human and assistant texts directly from the example
    human_text = example['text'].split('<HUMAN>:')[1].split('<ASSISTANT>:')[0].strip()
    assistant_text = example['text'].split('<ASSISTANT>:')[1].strip()

    # Apply the new template
    transformed_text = f'<s>[INST] {human_text} [/INST] {assistant_text} </s>'

    return {'instruction': human_text, 'output': assistant_text}

# Example text
# example_text = """<HUMAN>: What is a panic attack? <ASSISTANT>: Panic attacks come on suddenly and involve intense and often overwhelming fear. They’re accompanied by very challenging physical symptoms, like a racing heartbeat, shortness of breath, or nausea. Unexpected panic attacks occur without an obvious cause. Expected panic attacks are cued by external stressors, like phobias. Panic attacks can happen to anyone, but having more than one may be a sign of panic disorder, a mental health condition characterized by sudden and repeated panic attacks."""

# Transform the conversation
# transformed_text = transform_conversation({'text': example_text})
transformed_dataset = dataset.map(transform_conversation)
# print(transformed_text['text'])
```
